```python
import bittensor as bt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
m = bt.metagraph().sync() 
```

```python
m.dividends[0]
```

```python
index = m.uids.tolist()
columns = [ 'uid', 'active', 'stake', 'rank', 'trust', 'consensus', 'incentive', 'dividends', 'emission']
df = pd.DataFrame(columns = columns, index = index)
for uid in m.uids.tolist():
    v = {
        'uid': m.uids[uid].item(),
        'active': m.active[uid].item(),             
        'stake': m.stake[uid].item(),             
        'rank': m.ranks[uid].item(),            
        'trust': m.trust[uid].item(),             
        'consensus': m.consensus[uid].item(),             
        'incentive': m.incentive[uid].item(),             
        'dividends': m.dividends[uid].item(),             
        'emission': m.emission[uid].item()}
    df.loc[uid] = pd.Series(v)
df['uid'] = df.index
```

```python
df
```

```python
markers = {1: "s", 0: "X"}
fig, ax = plt.subplots(figsize=(20, 5))
ax.set_title('Dividends')
sns.scatterplot(
    data = df, 
    x = "uid", 
    y = "dividends", 
    size = "stake", 
    hue = 'stake', 
    hue_norm = (0, 1000), 
    style = "active", 
    markers = markers
)
```

```python
markers = {1: "s", 0: "X"}
fig, ax = plt.subplots(figsize=(20, 5))
ax.set_title('Consensus')
sns.scatterplot (
    data = df, 
    x = "uid", 
    y = "consensus", 
    size = "rank", 
    hue = 'stake', 
    hue_norm = (0, 1000), 
    style = "active", 
    markers = markers
)
```
